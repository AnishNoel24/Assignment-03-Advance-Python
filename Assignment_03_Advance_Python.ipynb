Q1. Create a file that contains 1000 lines of random strings.

Q5. Convert all the files of Q4 into upper case one by one.

```python
import os

file_sizes = [1, 2, 3, 4, 5]  # Sizes in GB

for size in file_sizes:
    file_name = f'random_strings_{size}GB.txt'
    output_file = f'random_strings_{size}GB_uppercase.txt'

    with open(file_name, 'r') as file:
        content = file.read()

    uppercase_content = content.upper()

    with open(output_file, 'w') as file:
        file.write(uppercase_content)

    print(f"File {output_file} created successfully!")

print("All files converted to uppercase!")
```

Q6. Convert all the files of Q4 into upper case parallel using multi-threading

```python
import os
import concurrent.futures

file_sizes = [1, 2, 3, 4, 5]  # Sizes in GB

def convert_to_uppercase(file_name):
    output_file = f'random_strings_{file_name}GB_uppercase.txt'

    with open(file_name, 'r') as file:
        content = file.read()

    uppercase_content = content.upper()

    with open(output_file, 'w') as file:
        file.write(uppercase_content)

    print(f"File {output_file} created successfully!")


with concurrent.futures.ThreadPoolExecutor() as executor:
    futures = []
    for size in file_sizes:
        file_name = f'random_strings_{size}GB.txt'
        future = executor.submit(convert_to_uppercase, file_name)
        futures.append(future)

    concurrent.futures.wait(futures)

print("All files converted to uppercase!")
```

Q10. Create an automated pipeline using multi-threading for:
“Automatic Download of 100 Videos from YouTube” → “Convert it to Audio”

```python
from pytube import YouTube
from moviepy.editor import VideoFileClip

# List of YouTube video URLs
video_urls = [
    "https://www.youtube.com/watch?v=video1",
    "https://www.youtube.com/watch?v=video2",
    "https://www.youtube.com/watch?v=video3",
    ...
]

# Output directory for downloaded videos and audio
output_directory = "videos"

# Function to download a video from YouTube
def download_video(url):
    yt = YouTube(url)
    video = yt.streams.filter(only_video=True).first()
    video.download(output_path=output_directory)

# Function to convert a video to audio
def convert_to_audio(video_file):
    video = VideoFileClip(video_file)
    audio_file = video_file.replace('.mp4', '.mp3')
    video.audio.write_audiofile(audio_file)

# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Download videos
for url in video_urls:
    download_video(url)

# Convert downloaded videos to audio
for video_file in os.listdir(output_directory):
    if video_file.endswith('.mp4'):
        convert_to_audio(os.path.join(output_directory, video_file))

print("Pipeline completed!")
```

Q11. Create an automated pipeline using multi-threading for: “Automatic Download of 500 images of Dog from
GoogleImages” → “Rescale it to 50%”.

```python
import os
import requests
import concurrent.futures
from PIL import Image


# Define the search query and the number of images to download
search_query = "dog"
num_images = 500

# Output directory for downloaded images
output_directory = "dog_images"


# Function to download an image from a given URL
def download_image(url, filename):
    response = requests.get(url, stream=True)
    response.raise_for_status()

    with open(filename, 'wb') as file:
        for chunk in response.iter_content(chunk_size=8192):
            file.write(chunk)


# Function to rescale an image to 50%
def rescale_image(filename):
    image = Image.open(filename)
    new_size = (image.width // 2, image.height // 2)
    resized_image = image.resize(new_size)
    resized_image.save(filename)


# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)


# Define a function to be executed by each thread
def process_image(index):
    try:
        # Generate a unique filename for each image
        filename = os.path.join(output_directory, f"image_{index}.jpg")

        # Download the image
        image_url = f"https://www.google.com/search?q={search_query}&tbm=isch&start={index}&sa=X&tbs=isz:l&tbm=isch"
        download_image(image_url, filename)

        # Rescale the image
        rescale_image(filename)

        print(f"Processed image {index}")
    except Exception as e:
        print(f"Error processing image {index}: {e}")


# Create a thread pool with a maximum of 10 threads
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    # Submit tasks to the thread pool
    image_indices = range(0, num_images)
    futures = [executor.submit(process_image, index) for index in image_indices]

    # Wait for all tasks to complete
    concurrent.futures.wait(futures)

print("Pipeline completed!")
```

Q12. Create a random dataset of 100 rows and 30 columns. All the values are defined between [1,200]. Perform
the following operations:
(i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number
rows having missing values.
(ii) Replace all the NA values with the average of the column value.
(iii) Find the Pearson correlation among all the columns and plot heat map. Also select those columns
having correlation <=0.7.
(iv) Normalize all the values in the dataset between 0 and 10.
(v) Replace all the values in the dataset with 1 if value <=0.5 else with 0.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Set the random seed for reproducibility
np.random.seed(42)

# Generate random dataset of 100 rows and 30 columns
dataset = pd.DataFrame(np.random.randint(1, 201, size=(100, 30)))

# Replace values with NA in the range [10, 60]
dataset.loc[10:60] = np.nan

# Count the number of rows with missing values
missing_rows_count = dataset.isnull().sum(axis=1).sum()
print(f"Number of rows with missing values: {missing_rows_count}")

# Replace NA values with the average of the column
dataset = dataset.fillna(dataset.mean())

# Calculate Pearson correlation among columns
correlation_matrix = dataset.corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Pearson Correlation Heatmap")
plt.show()

# Select columns with correlation <= 0.7
columns_to_select = correlation_matrix.columns[correlation_matrix.abs().lt(0.7).all()]
selected_columns = dataset[columns_to_select]

# Normalize values between 0 and 10
normalized_dataset = (selected_columns - selected_columns.min()) * 10 / (selected_columns.max() - selected_columns.min())

# Replace values <= 0.5 with 1 and values > 0.5 with 0
replaced_dataset = normalized_dataset.applymap(lambda x: 1 if x <= 0.5 else 0)

# Print the modified dataset
print(replaced_dataset)
```

Q13. Create a random dataset of 500 rows and 10 columns.
Columns 1 to 4 are defined between [-10, 10];
Columns 5 to 8 are defined between [10, 20];
Columns 9 to 10 are defined between [-100, 100].
Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric
graph using each algorithm.
(i) K-Mean clustering
(ii) Hierarchical clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import pairwise_distances


# Set the random seed for reproducibility
np.random.seed(42)

# Generate random dataset of 500 rows and 10 columns
dataset = pd.DataFrame(np.random.uniform(-10, 10, size=(500, 4)), columns=['col1', 'col2', 'col3', 'col4'])
dataset = pd.concat([dataset, pd.DataFrame(np.random.uniform(10, 20, size=(500, 4)), columns=['col5', 'col6', 'col7', 'col8'])], axis=1)
dataset = pd.concat([dataset, pd.DataFrame(np.random.uniform(-100, 100, size=(500, 2)), columns=['col9', 'col10'])], axis=1)

# Apply K-Means clustering
kmeans = KMeans()
inertia = []
num_clusters_range = range(1, 11)  # Range of number of clusters to try

for num_clusters in num_clusters_range:
    kmeans.n_clusters = num_clusters
    kmeans.fit(dataset)
    inertia.append(kmeans.inertia_)

# Plot the distance metric graph for K-Means clustering
plt.plot(num_clusters_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('K-Means Clustering - Optimal Number of Clusters')
plt.show()

# Apply Hierarchical clustering
agglomerative = AgglomerativeClustering()
linkage_distances = ['ward', 'complete', 'average', 'single']
distance_metrics = ['euclidean', 'manhattan', 'cosine']

for linkage in linkage_distances:
    for metric in distance_metrics:
        agglomerative.linkage = linkage
        agglomerative.affinity = metric
        distances = pairwise_distances(dataset, metric=metric)
        clustering = agglomerative.fit(distances)

        # Plot the distance metric graph for Hierarchical clustering
        plt.plot(range(1, dataset.shape[0]), clustering.distances_, marker='o')
        plt.xlabel('Number of Clusters')
        plt.ylabel('Distance')
        plt.title(f'Hierarchical Clustering - Linkage: {linkage}, Metric: {metric}')
        plt.show()
```

Q14. Create a random dataset of 600 rows and 15 columns. All the values are defined between [-100,100].
Perform the following operations:
(i) Plot scatter graph between Column 5 and Column 6.
(ii) Plot histogram of each column in single graph.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random dataset of 600 rows and 15 columns
dataset = pd.DataFrame(np.random.uniform(-100, 100, size=(600, 15)))

# Plot scatter graph between Column 5 and Column 6
plt.scatter(dataset.iloc[:, 4], dataset.iloc[:, 5])
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.title('Scatter Plot: Column 5 vs Column 6')
plt.show()

# Plot histogram of each column in a single graph
fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(12, 8))
fig.suptitle('Histogram of Each Column')
for i, ax in enumerate(axes.flatten()):
    ax.hist(dataset.iloc[:, i], bins=10, edgecolor='black')
    ax.set_title(f'Column {i+1}')
plt.tight_layout()
plt.show()

# Plot the box plot of each column in a single graph
fig, ax = plt.subplots(figsize=(10, 6))
ax.boxplot(dataset.values, vert=False, labels=range(1, 16))
ax.set_title('Box Plot of Each Column')
ax.set_xlabel('Column')
ax.set_ylabel('Value')
plt.show()
```

Q15. Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].
Perform the following operations:
(i) Perform t-Test on each column.
(ii) Perform Wilcoxon Signed Rank Test on each column.
(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random dataset of 500 rows and 5 columns
dataset = pd.DataFrame(np.random.uniform(5, 10, size=(500, 5)))

# Perform t-Test on each column
t_test_results = []
for col in dataset.columns:
    t_statistic, p_value = ttest_1samp(dataset[col], 7.5)  # Perform one-sample t-test against null hypothesis mean of 7.5
    t_test_results.append((col, t_statistic, p_value))

# Perform Wilcoxon Signed Rank Test on each column
wilcoxon_results = []
for col in dataset.columns:
    statistic, p_value = wilcoxon(dataset[col] - 7.5)  # Perform Wilcoxon Signed Rank Test against null hypothesis median of 7.5
    wilcoxon_results.append((col, statistic, p_value))

# Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
t_test_2samp_stat, t_test_2samp_pvalue = ttest_ind(dataset[2], dataset[3])  # Perform two-sample t-test
ranksums_stat, ranksums_pvalue = ranksums(dataset[2], dataset[3])  # Perform Wilcoxon Rank Sum Test

# Print the results
print("T-Test Results:")
for result in t_test_results:
    print(f"Column: {result[0]}, T-Statistic: {result[1]}, P-Value: {result[2]}")

print("\nWilcoxon Signed Rank Test Results:")
for result in wilcoxon_results:
    print(f"Column: {result[0]}, Statistic: {result[1]}, P-Value: {result[2]}")

print("\nTwo Sample T-Test and Wilcoxon Rank Sum Test Results:")
print(f"Column 3 vs Column 4 - T-Test: T-Statistic: {t_test_2samp_stat}, P-Value: {t_test_2samp_pvalue}")
print(f"Column 3 vs Column 4 - Wilcoxon Rank Sum Test: Statistic: {ranksums_stat}, P-Value: {ranksums_pvalue}")
```
